# Şeker hastalığı (Diyabet) nedir?

Çağın hastalıkları arasında en ön sıralarda yer alan diyabet (şeker hastalığı), ölümcül birçok hastalığın oluşumunda birinci sırada rol oynayan ve dünyanın her yerinde çok yaygın olarak görülen bir hastalık türüdür. Hastalığın tam adı olan Diabetes Mellitus, Yunancada şekerli idrar anlamına gelir. Sağlıklı bireylerde açlık kan glukoz düzeyi 70-100 mg/dL aralığındadır. Kandaki şeker seviyesinin bu aralığın üzerine çıkması, genellikle diyabet hastalığına işaret eder. Hastalığın nedeni insülin hormonu üretiminin herhangi bir nedenle yetersiz olması veya hiç olmaması, ya da vücut dokularının insüline karşı duyarsız hale gelmesidir. Birçok farklı çeşidi bulunan diyabet hastalığının en yaygın görülen ve genellikle 35-40 yaş üzeri bireylerde ortaya çıkan türü Tip 2 diyabettir. Bir diğer adı insülin direnci olan Tip 2 diyabette pankreasta insülin üretimi yeterli olduğu halde hücrelerde insülin hormonunu algılayıcı reseptörlerin çalışmaması nedeniyle bu hormona karşı duyarsızlık gelişir. Bu durumda kan şekeri insülin tarafından dokulara taşınamaz ve kan glukoz düzeyi normalin üzerine çıkar. Bu durum ağız kuruması, kilo kaybı, çok su içme ve çok yemek yeme şeklinde belirtiler ile kendini gösterir.

## 3-Integrated Noise Model (INM)

```python
from sklearn.neural_network import MLPClassifier
```

```python
from sklearn.neural_network import MLPRegressor
```

Burada yine diyabetleri tanıttık.

```python
diy = pd.read_csv('diabetes.csv')
df = diy.copy()
df = df.dropna()
y=df["Outcome"]
x=df.drop(["Outcome"],axis=1)

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

Eğiteceğimiz X verilerimizi normalize ediyoruz.Ve verilerimiz görüntülüyoruz.

```python
#değişken standartlaştırması
scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
mlp_model = MLPRegressor(hidden_layer_sizes = (15,6)).fit(X_train_scaled, y_train)
print(mlp_model)
print(mlp_model.n_layers_)
print(mlp_model.hidden_layer_sizes)
```

```python
X_test_scaled[0:5]
```

Burada yine bir tahmin modeli oluşturalım ve bu modelde mlp yani satır 79,80 deki MLPClassifier,MLPRegressor diye tanıttığımız keras modelinden çekiyoruz daha sonra bunu modele oturtuyoruz.

```python
y_pred = mlp_model.predict(X_test_scaled)
np.sqrt(mean_squared_error(y_test, y_pred))
```

Hemen modeli görüntülüyoruz.Model de bize 15 tane gizli katmanın olsun ve her gizli katmanda 6 nöron olsun diyor.

```python
mlp_model
```

Burada parametreler tanımlıyorum.aktivasyon fonksiyonu,ara katman seviyesi ve alfa katsayıları.Bunları bir sözlük içerisine aldım.Gördüğünüz gibi alpha anahtar değerler bunun Ki leri.

```python
mlp_params = {'alpha': [0.1, 0.01,0.02,0.005],
             'hidden_layer_sizes': [(18,11),(64,50,8),(128,64,32),(10,7,1)],
             'activation': ['relu','logistic']}
```

GridSearch ile aratıyorum.Bakalım en iyi parametre hangisi çıkacak.

```python
mlp_cv_model = GridSearchCV(mlp_model, mlp_params, cv = 10)
mlp_cv_model.fit(X_train_scaled, y_train)
```

Bu model için en iyi aktivasyon fonksiyon türü  logistic fonksiyon,alpha kat sayım 0.02,ara katman sayım da (128, 64, 32) olmalı 

```python
mlp_cv_model.best_params_
```

En iyi aktivasyon fonksiyonunu en iyi alpha modelini,en iyi ara katman sayısını yazdırdım.

```python
mlp_tuned = MLPRegressor(activation=mlp_cv_model.best_params_['activation'],
                         alpha = mlp_cv_model.best_params_['alpha'], 
                         hidden_layer_sizes = mlp_cv_model.best_params_['hidden_layer_sizes'])
print(mlp_tuned)
print(mlp_tuned.activation)
```

Dönüştürülmüş yapay sinir ağı modelini fit ediyorum.

```python
mlp_tuned.fit(X_train_scaled, y_train)
```

Daha sonra bu dönüştürülmüş yapay sinir ağı modelinindeki normalize ettğimiz test verileri yeni tahmin verilerine atıyoruz.Dönüştürdüğümüz modeldeki en iyi parametrelerin hata  kareleri ortalamasını vermesini istiyoruz ve 0.42 değeri çıkıyor.

```python
y_pred = mlp_tuned.predict(X_test_scaled)
np.sqrt(mean_squared_error(y_test, y_pred))
```
